```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Practice-Datasets.csv")
df.head(5)
```

1. Create a histogram to analyze the distribution of Age.

```python
plt.figure(figsize=(5,5))
sns.histplot(df["Age"],edgecolor='black',color='red',alpha=0.7,kde=True,bins=30)
plt.title("Distributions of Age")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

2. Plot a pie chart showing the percentage distribution of gender

```python
plt.figure(figsize=(5,5))
plt.pie(
    df["Gender"].value_counts(),
    labels=['Male','Female'],
    colors=['lightblue','lightcoral'],
    autopct='%1.1f%%',
    startangle=90,
    shadow=True,
    wedgeprops={'edgecolor':'black','linewidth':1},
    textprops={'fontsize':10,'color':'black'}
)
plt.title("Percentage Distribution of Gender")
plt.show()
```

3. Create a bar chart showing the number of customers in each City.

```python
city_counts = df["City"].value_counts()
plt.figure(figsize=(5,5))
city_counts.plot(kind="bar", color="green", edgecolor="black", width=0.5, alpha=0.8)
plt.title("Customer Count in each City")
plt.xlabel("City")
plt.ylabel("Customer count")
plt.show()
```

4. Plot a box plot to analyze Monthly Income distribution.

```python
plt.figure(figsize=(5,5))
sns.boxplot(df["Monthly_Income"].dropna(),color="pink",width=0.5)
plt.title("Monthly Income Distribution")
plt.show()
```

5. Create a histogram of Spending Score and interpret the skewness.

```python
plt.figure(figsize=(5,5))
sns.histplot(df["Spending_Score"],edgecolor='black',alpha=0.7,kde=True,bins=50)
plt.title("Distributions of Spending Score")
plt.xlabel("Spending Score")
plt.show()
print("Skewness:", df["Spending_Score"].skew())
```

6. Plot a line chart to visualize Purchase Amount trend over Purchase Date.

```python
df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])
df["Month"] = df["Purchase_Date"].dt.month
monthly_sales = df.groupby("Month")["Purchase_Amount"].sum()
plt.figure(figsize=(8,5))
monthly_sales.plot(color='green',marker="o",linewidth=2)
plt.title("Monthly Purchase Trend")
plt.xlabel("Month")
plt.ylabel("Purchase Amount")
plt.grid(True)
plt.show()
```

7. Create a bar plot to compare average Monthly Income by Gender.

```python
avg_income_gender = df.groupby("Gender")["Monthly_Income"].mean()
plt.figure(figsize=(5,5))
avg_income_gender.plot(kind="bar",color='purple',width=0.5,edgecolor='black',alpha=0.6)
plt.title("Average Monthly Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Monthly Income")
plt.show()
```

8. Plot a box plot of Purchase Amount grouped by Product Category.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Product_Category"],y=df["Purchase_Amount"],color='blue',width=0.5)
plt.title("Amount on each Category of Product")
plt.xlabel("Product category")
plt.ylabel("Amount")
plt.show()
```

9. Create a pie chart to show Product Category distribution.

```python
plt.figure(figsize=(5,5))
plt.pie(
    df["Product_Category"].value_counts(),
    labels=["Clothing","Sports","Groceries","Electronics","Furniture"],
    colors=['red','green','blue','orange','purple'],
    autopct="%1.1f%%",
    startangle=90,
    shadow=True,
    wedgeprops= {'edgecolor':'black','linewidth':2}, 
    textprops= {'fontsize':12,'color':'black'}
)
plt.title("Product Category Distribution")
plt.show()
```

10. Plot a histogram of Purchase Amount with appropriate bin size.

```python
plt.figure(figsize=(8,5))
sns.histplot(df["Purchase_Amount"],kde=4,color='pink',bins=30,alpha=True,edgecolor='black')
plt.title("Distribution of Purchase Amount")
plt.xlabel("Purchase Amount")
plt.ylabel("Frequency")
plt.show()
```

11. Create a bar chart showing average Spending Score by City.

```python
df["Spending_Score"]=pd.to_numeric(df["Spending_Score"],errors="coerce")
city_avg=df.groupby("City")["Spending_Score"].mean().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x="City",y="Spending_Score",data=city_avg,color="yellow")
plt.title("Average Spending Score by City")
plt.xlabel("City")
plt.ylabel("Average Score")
plt.show()
```

12. Plot a line chart showing total Purchase Amount per month.

```python
df["Purchase_Date"] = pd.to_datetime(df["Purchase_Date"])
df["Month"] = df["Purchase_Date"].dt.month
monthly_sales = df.groupby("Month")["Purchase_Amount"].sum()
plt.figure(figsize=(8,5))
monthly_sales.plot(color='orange',marker="*",linewidth=2)
plt.title("Monthly Purchase Trend")
plt.xlabel("Month")
plt.ylabel("Purchase Amount")
plt.grid(True)
plt.show()
```

13. Create a box plot comparing Monthly Income across different Cities.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["City"],y=df["Monthly_Income"],color='brown',width=0.5)
plt.title("Monthly Income by City")
plt.xlabel("Product category")
plt.ylabel("Amount")
plt.show()
```

14. Plot a histogram to analyze Age distribution by Gender.

```python
plt.figure(figsize=(8,5))
sns.histplot(df[df["Gender"]=="Male"]["Age"],color="orange",label="Male",bins=20, alpha=0.5)
sns.histplot(df[df["Gender"]=="Female"]["Age"],color="green",label="Female",bins=20,alpha=0.5)
plt.title("Age Distribution by Gender")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

15. Create a bar chart showing total Purchase Amount by Product Category.


```python
df["Purchase_Amount"]=pd.to_numeric(df["Purchase_Amount"],errors="coerce")
cat_sum=df.groupby("Product_Category")["Purchase_Amount"].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(x="Product_Category",y="Purchase_Amount",data=cat_sum,color="red",alpha=0.5)
plt.title("Amount by Product Category")
plt.xlabel("Product Category")
plt.ylabel("Purchase Amount")
plt.show()
```

16. Plot a scatter plot between Monthly Income and Spending Score.


```python
plt.figure(figsize=(8,5))
sns.scatterplot(x=df['Monthly_Income'],y=df['Spending_Score'],color='purple',edgecolor='black',alpha=0.7)
plt.title("Monthly Income vs Spending Score")
plt.xlabel("Monthly Income")
plt.ylabel("Score")
plt.show()
```

17. Create a line chart to visualize daily Purchase Amount variation.

```python
df["Purchase_Amount"] = pd.to_numeric(df["Purchase_Amount"],errors="coerce")
daily_variation = df.groupby("Purchase_Date")["Purchase_Amount"].sum()
plt.figure(figsize=(8,5))
sns.lineplot(daily_variation,color='orange',marker="o",linewidth=2)
plt.title("Daily Purchase Amount Variation")
plt.xlabel("Amount")
plt.ylabel("Frequency")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

18. Plot a box plot of Spending Score grouped by Gender.

```python
plt.figure(figsize=(8,5))
sns.boxplot(x=df["Gender"],y=df["Spending_Score"],color='green',width=0.5)
plt.title("Spending Score grouped by Gender")
plt.xlabel("Gender")
plt.ylabel("Spending Score")
plt.show()
```

19. Create a bar chart showing customer count by Product Category and Gender.

```python
count_data = df.groupby(["Product_Category", "Gender"]).size().reset_index(name="Customer_Count")
plt.figure(figsize=(10,6))
sns.barplot(x="Product_Category",y="Customer_Count",hue="Gender",data=count_data)
plt.title("Customer Count by Product Category and Gender")
plt.xlabel("Product Category")
plt.ylabel("Customer Count")
plt.show()
```
